

Data Preparing

```python
import pandas as pd
dataset = pd.read_csv('HousingData.csv')
```

```python

dataset
```

```python
dataset.dropna()
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isnull().sum()
```

```python
import numpy as np
dataset.replace(to_replace = np.NaN, value = -99)
dataset[:468]
```

```python
dataset = pd.concat([dataset['MEDV'],dataset.drop(['MEDV'], axis=1)],axis=1)
```

```python
from sklearn.model_selection import train_test_split
training_dataset, validation_dataset = train_test_split(dataset, test_size=0.1)
```

```python
training_dataset.to_csv('training_dataset.csv',index=False, header=False)
validation_dataset.to_csv('validation_dataset.csv',index=False, header=False)
```

```python
import sagemaker
sess = sagemaker.Session()
bucket = sess.default_bucket()

role = sagemaker.get_execution_role()
print(role)
```

```python
prefix = 'boston-housing'
training_data_path = sess.upload_data(
 path='training_dataset.csv',
 key_prefix=prefix + '/input/training')
validation_data_path = sess.upload_data(
 path='validation_dataset.csv',
 key_prefix=prefix + '/input/validation')
print(training_data_path)
print(validation_data_path)
```

Configuring training job

```python
import boto3
from sagemaker import image_uris
region = boto3.Session().region_name
container = image_uris.retrieve('linear-learner', region)
```

```python
import boto3
import sagemaker

sess = sagemaker.Session()

linear = sagemaker.estimator.Estimator(
    container,
    role=role,
    instance_count=1,
    instance_type="ml.m4.xlarge",
    output_path='s3://{}/{}/output'.format(bucket,prefix),
    sagemaker_session=sess)
```

```python
linear.set_hyperparameters(
 predictor_type='regressor', 
 mini_batch_size=32)
```

```python
training_data_channel = sagemaker.TrainingInput(s3_data=training_data_path,content_type='text/csv')
```

```python
validation_data_channel = sagemaker.TrainingInput(s3_data=validation_data_path,content_type='text/csv')
```

```python
linear.fit({'train': training_data_channel,'validation': validation_data_channel})
```

Model Deploy

```python
linear_regressor = linear.deploy(initial_instance_count = 1,
                                          instance_type = 'ml.m4.xlarge')

#need to make sure data is in correct format for deployed model
from sagemaker.predictor import csv_serializer, csv_deserializer
linear_regressor.serializer = csv_serializer
linear_regressor.deserializer = csv_deserializer
```

```python
linear_regressor.ContentType = 'text/csv'
linear_regressor.Accept = 'text/csv'
linear_regressor.serializer =sagemaker.serializers.CSVSerializer()
linear_regressor.deserializer =sagemaker.deserializers.CSVDeserializer()
test_sample = '0.00632,	18.0,	2.31,	0.0,	0.538,	6.575,	65.2,	4.0900,	1,	296,	15.3,	396.90,	4.98'
response = linear_regressor.predict(test_sample)
print(response)
```

```python
linear_regressor.delete_endpoint()
```

```python
import boto3
from sagemaker import image_uris
region = boto3.Session().region_name
container = image_uris.retrieve('xgboost', region,version='latest')
```

```python
xgb_estimator = sagemaker.estimator.Estimator(container,
 role=sagemaker.get_execution_role(),
 instance_count=1,
 instance_type='ml.m4.xlarge',
 output_path='s3://{}/{}/output'.format(bucket,
 prefix))
```

```python
xgb_estimator.set_hyperparameters(
 objective='reg:linear',
 num_round=200,
 early_stopping_rounds=10)
```

```python
training_data_channel = sagemaker.TrainingInput(s3_data=training_data_path,content_type='text/csv')
validation_data_channel = sagemaker.TrainingInput(s3_data=validation_data_path,content_type='text/csv')
```

```python
xgb_estimator.fit({'train': training_data_channel,
 'validation': validation_data_channel})
```

```python
from time import strftime, gmtime
timestamp = strftime('%d-%H-%M-%S', gmtime())
endpoint_name = 'xgb-demo'+'-'+timestamp
xgb_predictor = xgb_estimator.deploy(
 endpoint_name=endpoint_name,
 initial_instance_count=1,
 instance_type='ml.t2.medium')
```

```python
test_sample = '0.00632,18.00,2.310,0,0.5380,6.5750,65.20,4.0900,1,296.0,15.30,4.98'
xgb_predictor.ContentType = 'text/csv'
xgb_predictor.serializer =sagemaker.serializers.CSVSerializer()
xgb_predictor.deserializer =sagemaker.deserializers.CSVDeserializer()
response = xgb_predictor.predict(test_sample)
print(response)
```

```python
xgb_predictor.delete_endpoint()
```
